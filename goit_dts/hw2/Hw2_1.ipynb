Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання


```python
import pandas as pd
# Зчитуємо таблицю за посиланням за записуємо по змінної
df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match = "Коефіцієнт народжуваності в регіонах України") 
# Залишаємо вхідні дані без змін, та працюємо надалі з копією
df1 = pd.DataFrame(df[0])
df1
```


-Вивести перші рядки таблиці за допомогою методу *head*

```python
df1.head(5)
```

-Визначте кількість рядків та стовпців у датафреймі (атрибут shape)


```python
df1.shape
```

-Замініть у таблиці значення "—" на значення NaN

```python
df1 = df1.replace("—", "NaN")
df1
```

-Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df1.dtypes
```

-Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
# За попередньо виявленими стовбчиками з невірним типом object, замінюємо по черзі
df1["2014"] = df1["2014"].astype(float)
df1["2019"] = df1["2019"].astype(float)
# Перевіряємо результат
df1.dtypes
```

-Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
#Кількість пропусків
res1 = df1.isnull().sum()
print("Кількість пропусків")
print(res1)
#Частка пропусків
res2 = df1.isna().mean()
print("\nЧастка пропусків")
print(res2)
```

-Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df1.drop(df1.index[-1], inplace=True)
df1
```

-Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)


```python
df1.fillna(df1.select_dtypes(include="number").mean(), inplace=True)
df1
```

-Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні


```python
mean_2019 = df1["2019"].mean()

df1[df1["2019"]> mean_2019]["Регіон"].tolist()
```

-У якому регіоні була найвища народжуваність у 2014 році?


```python
max_pop = df1["2014"].max()
df1[df1["2014"] == max_pop] ["Регіон"].values[0]
```

-Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
import matplotlib.pyplot as plt

%matplotlib inline

# Вибір даних для побудови діаграми
data = df1[['Регіон', '2019']]

# Побудова стовпчикової діаграми
plt.figure(figsize=(12, 6))  # Задання розміру діаграми
plt.bar(data['Регіон'], data['2019'])  # Побудова стовпчикової діаграми
plt.xlabel('Регіон')  # Підпис для осі x
plt.ylabel('Народжуваність')  # Підпис для осі y
plt.title('Народжуваність по регіонах у 2019 році')  # Заголовок діаграми
plt.xticks(rotation=90)  # Поворот підписів на осі x для зручності
plt.tight_layout()  # Оптимізація розміщення елементів на діаграмі
plt.show()  # Відображення діаграми
```

- Додайте від 3 до 5 графіків функцій різного типу на ваш вибір. 
Задайте графікам оформлення, щоб кожен графік у своїй домашній роботі чимось відрізнявся і не був схожим на інші. 
Можна використовувати як matplotlib, так і seaborn.

Не забудьте в Jupyter файл додати директиву %matplotlib inline, щоб графіки будувалися всередині документа.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
data = df1[['Регіон', '2014']]
labels = data['Регіон']


plt.pie(
    data['2014'],
    labels = labels,
    shadow=True,
    autopct="%.1f%%",
    pctdistance=0.7,
)
ax.set_title("Народжуваність по регіонам у 2014 році")
plt.show()
```
